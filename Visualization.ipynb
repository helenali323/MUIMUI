```python
insta_path = 'insta/'
product_path = 'data/'
```

```python
import seaborn as sns
import math
from PIL import Image
import cv2
import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = [15.0, 6.0]
def multi_plot(insta, fnames, ncols=3):
    """
        Display multiple images in a grid structure
        fnames: list of file name with full/relative path
    """
    assert ncols != 0
    no_imgs = len(fnames)+1
    nrows = math.ceil(no_imgs / ncols)
    fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(12, 3 * nrows),
                             subplot_kw={'xticks': [], 'yticks': []})
    fig.subplots_adjust(hspace=0.05, wspace=0.05)
    
    try:
        im = Image.open(insta_path+str(insta)+'.jpg')
        axes.flat[0].imshow(im)
        axes.flat[0].set_title(insta+'_insta', size=20)
    except:
        pass
    
    for ax, fname in zip(axes.flat[1:], fnames):
        try:
            im = Image.open(product_path+str(fname)+'.jpg')
            ax.imshow(im)
            ax.set_title(fname, size=20)
        except Exception as e:
            m = cv2.imread(product_path+str(fname)+'.jpg')
            ax.imshow(im)
            ax.set_title(fname, size=20)
    plt.tight_layout()
    plt.show()
```
